Function to read in relevant house ids

```python
def load_readings(houseid):
    'Funciton to load in data for the house electricity use and resample to 1 hour'
    
    data = ideal.get(homeid=houseid, subtype='electric-combined')[0]['readings']
    
    df = pd.DataFrame(data).resample('H').agg(lambda x: sum(x/3600)/1000)
    
    return df    
```

```python
def load_data(arr):
    'Function to load in all readings of relevant homes'
    
    data = {}
    
    for i in arr:
        
        print('i: ', i)

        data[i] = load_readings(int(i))
        
    return data
```
